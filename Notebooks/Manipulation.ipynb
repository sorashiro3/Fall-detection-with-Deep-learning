# Smote balancing

```python
import pandas as pd
import os

import warnings 
warnings.filterwarnings('ignore')

import glob
```

```python
df = []

col = ['TimeStamp(s)', 'FrameCounter', 'EulerX', 'EulerY', 'EulerZ']

for i in range(6, 10) :
    paths = glob.glob(f'sensor_data\SA0{i}\*.csv')
    
    for p in paths :
        tmp = pd.read_csv(p)
        tmp = tmp.drop(columns = col)
        
        df.append(tmp)

for i in range(10, 39) :
    paths = glob.glob(f'sensor_data\SA{i}\*.csv')
    
    for p in paths :
        tmp = pd.read_csv(p)
        tmp = tmp.drop(columns = col)

        df.append(tmp)

un_bal = pd.concat(df, ignore_index = True)
```

```python
df = []

col = ['TimeStamp(s)', 'FrameCounter', 'EulerX', 'EulerY', 'EulerZ']

for i in range(6, 10) :
    paths = glob.glob(f'sensor_data\SA0{i}Fall\*.csv')
    
    for p in paths :
        tmp = pd.read_csv(p)
        tmp = tmp.drop(columns = col)
        
        df.append(tmp)

for i in range(10, 39) :
    paths = glob.glob(f'sensor_data\SA{i}Fall\*.csv')
    
    for p in paths :
        tmp = pd.read_csv(p)
        tmp = tmp.drop(columns = col)
        
        df.append(tmp)

un_bal_Fall = pd.concat(df, ignore_index = True)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote_technique = SMOTE(sampling_strategy='auto')

balanced, balanced_Fall = smote_technique.fit_resample(un_bal, un_bal_Fall)
```
